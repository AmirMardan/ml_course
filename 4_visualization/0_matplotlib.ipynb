Let's have our first plot.

````python
# Fake up data for our plots
x1 = np.arange(-3 * np.pi, 3 * np.pi, np.pi/6)
x2 = np.arange(-3 * np.pi, 3 * np.pi, np.pi/3)

y1 = np.sin(x1)
y2 = np.cos(x2)

plt.plot(x1, # Values on x-axis
         y1, # Values on y-axis
         label='sin' # Label of the plot
         )

# Add title
plt.title('Simple plot')

# Add label to axis
plt.xlabel('X')
plt.ylabel('Y')

# Add grids
plt.grid()

# Add another plot
plt.plot(x2, y2, label='cos')

# Add legend
plt.legend()
````

Let's set the color of lines, markers, line style, and line width.

````python
# Let's add more parameters to our plot
plt.plot(x1, y1,
         label='sin',   # Label of the plot
         color='teal',  # Color of the line
         marker='v'     # Shape of the marker
         )    

# Add title
plt.title('Simple plot')

# Add label
plt.xlabel('X')
plt.ylabel('Y')

# Add grids
plt.grid()

# Add another plot
plt.plot(x2, y2, label='cos', color='plum', marker='o',
         linestyle='--',    # Line style
         linewidth=2        # Width of the line
         )

# Add legend
plt.legend(['Cosine', 'Sine'], # Labels
           loc='upper left'     # Location of the legend
           )    

# Change the y-ticks
plt.yticks(np.arange(-1, 1, 0.5))
````

### 1.2 Subplots

We can create a figure with subplots using `plt.subplot(n_row, n_col, number)`.

````python
# Simple subplot

plt.subplot(1, 2, 1)
plt.plot(x1, y1)
plt.title('Plot 1')

plt.subplot(1, 2, 2)
plt.plot(x2, y2)
plt.title('Plot 2')

plt.suptitle('Simple subplot')
````

We can also have one plot over multiple subplots!

````python
# Plot on multiple subplots 

plt.subplot(2, 1, 1)
plt.plot(x1, y1)
plt.title('Plot 1')

plt.subplot(2, 2, 3)
plt.plot(x2, y2)
plt.title('Plot 2')

plt.subplot(2, 2, 4)
plt.plot(x2, y2)
plt.plot(x1, y1, )
plt.title('Plot 3')

plt.suptitle('Simple subplot')
````

**Text and annotation**

To add more details to your plots, texts and annotations can be added.

````python
# Let's add a text to a plot with specific color and make it bold and italic

arrowprops = dict(arrowstyle="->",  
                  connectionstyle="arc3,rad=.3", # Specify the connection style
                  color='red',  # Set the color
                  lw=1  # Set the width of line
                  )

plt.plot(x2, y2)
plt.annotate('Local max', # text to annotate
             xy=(0., 1), # the position of x and y to point
             xytext=(1.5,.5), # the position of text
             color='brown', # color of text
             arrowprops=arrowprops, # property of the arrow
             fontsize=13, fontweight='bold', fontstyle='italic')


plt.annotate('Local min', # text to annotate
             xy=(-3.1,-1), 
             xytext=(-5, -.2), # the xy position to add the texts
             bbox=dict(boxstyle='round', fc='yellow'), # or property of text box
             arrowprops=dict(arrowstyle='wedge, tail_width=0.5', alpha=0.3));
````

### 1.3 Object-oriented method 

Let's level up the quality of our figures by using the object-oriented property of Matplotlib

````python
# Let's define a font style for title of plots first 
title_fontdic= {'size':9, 'color':'red', 'fontname':'Segoe Print', 'family':'serif'}


# Let's create a figure object

fig = plt.figure(figsize=(5, 2), 
           dpi=100, 
           facecolor='yellow' 
           )

ax1 = fig.add_subplot(1, 2, 1) # adding a subplot to figure object
ax1.plot(x1, y1)  # plot using ax object

ax1.set_facecolor('#E2D784') # set the face color for axis

ax1.set_xlabel('X')
ax1.set_ylabel('Y', labelpad=-10) # Decreasing the pad of label
ax1.set_title('Plot 1 \U0001F60A', loc='left', fontsize=10)
ax2 = fig.add_subplot(1, 2, 2) # adding another subplot to figure object
ax2.plot(x2, y2, 'lime')  # plot using ax object

# Set the face color for axis
ax2.set_facecolor('#F68989') 

# Set xlabel
ax2.set_xlabel('X')

# Deleting the y-ticks for second matrix
ax2.set_yticks([])  

# Set title for second plot
ax2.set_title('plot 2', loc='right', fontdict=title_fontdic) 

plt.subplots_adjust(wspace=0.1)  # reduce the space between two axis
````

Reference for <a href='https://matplotlib.org/stable/tutorials/text/text_props.html'>Text properties</a>

We can also use `ax.set_xlim()` and `ax.set_ylim()` to limit the axis.

We use `alpha` in a plot to make the line transparent.

````python
fig = plt.figure(figsize=(3, 2), 
           dpi=100, 
           facecolor='yellow' 
           )

ax1 = fig.add_subplot(1, 1, 1) # adding a subplot to figure object
ax1.plot(x1, y1, 'k')  # plot using ax object
ax1.plot(x2, y2, 'k',
         alpha=.2  # Alpha is used to create a transparent line
         )
ax1.set_xlim([-5, 5]) # Set limit for x-axis

ax1.grid()

````

### 2.1 Scatter plot

Scatter plots are used to see the relationship between two variables.

**Syntax**
```Python
matplotlib.pyplot.scatter(x, y, s=None, c=None, marker=None, cmap=None, norm=None, vmin=None, vmax=None, alpha=None, linewidths=None, *, edgecolors=None, plotnonfinite=False, data=None, **kwargs)
```

````python
fig = plt.figure(figsize=(3, 2))

ax = fig.add_subplot(1, 1, 1)
ax.scatter(x=x, # Data along x-direction
           y=y  # Data along x-direction
           );
````

We can set a color for each sample based on their category.

````python
fig = plt.figure(figsize=(3, 2))

ax = fig.add_subplot(1, 1, 1)
im = ax.scatter(x = x, # Data along x-direction
           y = y,  # Data along x-direction
           c = category,  # Set the color of markers based on an array
           s = importance,  # Set the size of markers based on an array
           )

ax.set_xlim([0, 1.5])
leg = ax.legend(*im.legend_elements(), # Specify where to find the labels
                title='Class', # Title of the legend
                fontsize=6, title_fontsize=7 # Font size of classes and title
                )


````

We can change the labels in legend using `set_text()`.

````python
fig = plt.figure(figsize=(3, 2))

ax = fig.add_subplot(1, 1, 1)
im = ax.scatter(x = x, # Data along x-direction
           y = y,  # Data along x-direction
           c = category,  # Set the color of markers based on an array
           s = importance,  # Set the size of markers based on an array
           )

ax.set_xlim([0, 1.7])
leg = ax.legend(*im.legend_elements(), # Specify where to find the labels
                title='Class', # Title of the legend
                fontsize=7, title_fontsize=9 # Font size of classes and title
                )


# Let's set the legend based on the following list
labels=['apple', 'banana', 'raisin', 'orange', 'strawberry']


[leg.texts[i].set_text(labels[i]) for i in range(len(labels))];
````

<a name='bar'></a>
### 2.2 Bar plot

Bar plots are used to study the categorical variable vs a numerical variable.

**Syntax**

```Python
matplotlib.pyplot.bar(x, height, width=0.8, bottom=None, *, align='center', data=None, **kwargs)
```

````python
# Let's create a dataset
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
ordered = [30, 50, 12, 62, 80]

error = 2 + 3 * np.random.rand(len(days))
````

````python
fig = plt.figure(figsize=(4, 4))

# Instead of seperating the number of rows, cols nad plot number, we can put them without seperator
ax= fig.add_subplot(111) 

# A simple bar plot
ax.bar(x=days, # Categorical variable
       height=ordered # Numeric variable
       )
````

````python
fig = plt.figure(figsize=(4, 4))
ax= fig.add_subplot(111) 

im = ax.bar(x=days,
       height=ordered, # Numeric variable
       width=0.5,  # Width of bars
       color='#39AEA9', # Color of face
       edgecolor='#062C30', # Color of edge
       alpha=.9, # Make the bars transparent 
       yerr=error # Show the error along y-axis
       )

# Set limit for y-axis
ax.set_ylim([0, max(ordered) + 20])

# Find label based on value of order
labels = np.full(len(ordered), 'ðŸ˜')
labels[np.array(ordered)>75] = 'ðŸ˜Š'
labels[np.array(ordered)<25] = 'ðŸ˜²'

# Set the new labels
ax.bar_label(im, labels=labels,
             padding=1, fontsize=20);
````

### 2.3 Histogram

Histograms are used to show the frequency occurrence of data.

**Syntax**
```Python
matplotlib.pyplot.hist(x, bins=None, range=None, density=False, weights=None, cumulative=False, bottom=None, histtype='bar', align='mid', orientation='vertical', rwidth=None, log=False, color=None, label=None, stacked=False, *, data=None, **kwargs)
```

````python
# Simple histogram
fig = plt.figure(figsize=(3, 3))

# Instead of seperating the number of rows, cols nad plot number, we can put them without seperator
ax= fig.add_subplot(111) 

# A simple hist plot
plt.hist(random_letter1);
````

````python
fig = plt.figure(figsize=(3, 3))

# Instead of seperating the number of rows, cols nad plot number, we can put them without seperator
ax= fig.add_subplot(111) 

ax.hist(random_letter1,
         color='#39AEA9', # Color of face
         edgecolor='#062C30', # Color of edge
         orientation='horizontal', # Horizontal plot
         bins=range(5), # equal-width bins
         align='left');  
````

**Hist for more than one variable**

In this case, the `histtype` is important.

````python
fig = plt.figure(figsize=(5, 3))

# Instead of seperating the number of rows, cols nad plot number, we can put them without seperator
ax= fig.add_subplot(221) 
ax.hist([random_letter1, random_letter2],
         color=['#39AEA9', '#F1E1A6'], # Color of face
         edgecolor='#062C30', # Color of edge
         orientation='vertical', # Vertical plot
         bins=range(5), # equal-width bins
         histtype='bar',
         align='left'); 
ax.set_title("bar")

ax= fig.add_subplot(222) 
ax.hist([random_letter1, random_letter2],
         color=['#39AEA9', '#F1E1A6'], # Color of face
         edgecolor='#062C30', # Color of edge
         orientation='vertical', # Vertical plot
         bins=range(5), # equal-width bins
         histtype='barstacked',
         align='left'); 
ax.set_title("barstacked")

````

**2D histogram**

````python
# Generate data
mean = [0, 0]
cov = [[1, 1], [1, 2]]

x_hsit2d, y_hsit2d = np.random.multivariate_normal(mean, cov, 10000).T

fig = plt.figure(figsize=(12, 3))

ax= fig.add_subplot(131) 
ax.hist(x=x_hsit2d)
ax.set_title('Variable x')

ax= fig.add_subplot(132) 
ax.hist(x=y_hsit2d)
ax.set_title('Variable y')

ax= fig.add_subplot(133) 
ax.hist2d(x=x_hsit2d, y=y_hsit2d,
         cmap='jet');
ax.set_title('2D histogram')
````

### 2.5 Box Plot

A boxplot is used to display the distribution of data based on a five-number summary:
- minimum
- the first quartile (Q1)
- median
- the third quartile (Q3)
- maximum
Box plots can help us to find outliers.

```

     Q1-1.5IQR   Q1   median  Q3   Q3+1.5IQR
                  |-----:-----|
  o      |--------|     :     |--------|    o  o
                  |-----:-----|
outlier           <----------->            outliers
                       IQR

```

**Syntax**
```Python
matplotlib.pyplot.boxplot(x, notch=None, sym=None, vert=None, whis=None, positions=None, widths=None, patch_artist=None, bootstrap=None, usermedians=None, conf_intervals=None, meanline=None, showmeans=None, showcaps=None, showbox=None, showfliers=None, boxprops=None, labels=None, flierprops=None, medianprops=None, meanprops=None, capprops=None, whiskerprops=None, manage_ticks=True, autorange=False, zorder=None, *, data=None)
```

````python
# Let's  generate some data
spread = np.random.rand(50) * 100
center = np.ones(25) * 50
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100
data = np.concatenate((spread, center, flier_high, flier_low))
````

Here I plot the histogram as well to show how much data exist between Q1 and Q3.

````python
# A simple plot

fig = plt.figure(figsize=(6, 3))
ax= fig.add_subplot(121) 
ax.boxplot(data)
ax.set_title('Box plot')


ax= fig.add_subplot(122) 
ax.set_title('Hist plot')
ax.hist(data);
````

````python
fig = plt.figure(figsize=(3, 3))
ax= fig.add_subplot(111) 
ax.boxplot(data, 
           notch=True, # To put a notch on median
           sym='x', # Marker for outliers
           vert=False, # Plot horizontal
           patch_artist=True,
           capprops=dict(  # Property of caps
               color = '#062C30',  # Color
               lw = 2.5, # Line width
           ),
           whiskerprops = dict( # Property of whisker
               color = '#39AEA9',   # Color
               lw = 1.5,  # Line width
               ),
            medianprops = dict(  # Property of median line
                color = 'blue',  # Color
                lw = 1.5    # Line width
                ),
            boxprops = dict( # Properties of the box
                facecolor = '#F1E1A6', # Background color
                edgecolor = '#062C30',  # COlor line
                lw = 1.5  # Line width
                ),
           );
````

<a name='violin'></a>
### 2.6 Violin plot

A violin plot is more informative than a plain box plot. While a box plot only shows summary statistics such as mean/median and interquartile ranges, the violin plot shows the full distribution of the data.

**Syntax**
```Python
matplotlib.pyplot.violinplot(dataset, positions=None, vert=True, widths=0.5, showmeans=False, showextrema=True, showmedians=False, quantiles=None, points=100, bw_method=None, *, data=None)
```

As a simple example, I put violin and box plots next to each other for a better comparison.

````python
# A simple plot

fig = plt.figure(figsize=(6, 3))
ax= fig.add_subplot(121) 
ax.violinplot(data)
ax.set_title('Violin plot')


ax= fig.add_subplot(122) 
ax.set_title('Box plot')
ax.boxplot(data);
````

````python
fig = plt.figure(figsize=(3, 3))
ax= fig.add_subplot(111) 
ax.violinplot(data, 
           vert=False, # Plot horizontal
           showmeans=False, # To show the mean
           showextrema=True,  # To show the extrema
           showmedians=True,  # To show the median
           );
````

So picking a good <a href='https://matplotlib.org/stable/tutorials/colors/colormaps.html'>colormap</a> and color limit is very important if you need to compare the values of two 2D variables.

## 4. Animation using Matplotlib

To make animation using Matplotlib, you need to import another module from this package as

```Python
import matplotlib.animation as FuncAnimation 
```

````python
from matplotlib.animation import FuncAnimation

# Enable interactive plot
%matplotlib notebook

fig = plt.figure()
ax = plt.axes(xlim=(0, 5), ylim=(-2, 2))
line, = ax.plot([], [], lw=3)


def init():
    line.set_data([], [])
    return line, 

    
def animate(i):
    x = np.linspace(0, 5, 1000)
    f = 1
    y = np.sin(2 * f * np.pi * x + i)
    line.set_data(x, y)
    return line,

    
anim = FuncAnimation(fig, animate, init_func=init,
                     frames=200, interval=40, blit=True)
plt.show()
anim.save('./img/animation.gif', writer='imagemagick', fps=60)

    
````

### 4.1 Live graph with Matplotlib

Here is an example of live plotting. There is a text file in the repository called `stock.txt`. By changing this file, the plot gets updated automatically.

````python
fig = plt.figure()
#creating a subplot 
ax1 = fig.add_subplot(1,1,1)

def animate(i):
    data = open('./stock.txt','r').read()
    lines = data.split('\n')
    xs = []
    ys = []
   
    for line in lines:
        x, y = line.split(',') # Delimiter is comma    
        xs.append(float(x))
        ys.append(float(y))
   
    
    ax1.clear()
    ax1.plot(xs, ys)

    plt.xlabel('Day of trade')
    plt.ylabel('Price')
    plt.title('Live graph with Matplotlib')

    
ani = FuncAnimation(fig, animate, interval=1000) 
plt.show()
````

By changing `stock.txt`, we should see the changes live on the plot
